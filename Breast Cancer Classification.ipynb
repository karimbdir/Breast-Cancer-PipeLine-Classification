```python
X = df.drop(['id','diagnosis','Unnamed: 32'], axis=1)
```

```python
from sklearn.preprocessing import LabelEncoder
df['diagnosis'] = LabelEncoder().fit_transform(df['diagnosis'])
y= df['diagnosis']
```

```python
X.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.3, random_state=1)
```

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
```

```python
pipeline = Pipeline([
    ('Scaler', StandardScaler()), 
    ('clf', LogisticRegression()) 
])
pipeline.steps
```

```python
from sklearn.model_selection import cross_validate
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import AdaBoostClassifier,BaggingClassifier,RandomForestClassifier,GradientBoostingClassifier
from sklearn.neural_network import MLPClassifier


clfs = []
clfs.append(LogisticRegression())
clfs.append(MLPClassifier(solver='lbfgs', alpha=1e-5, hidden_layer_sizes=(150, 10)))
clfs.append(KNeighborsClassifier(n_neighbors=5))
clfs.append(DecisionTreeClassifier())
clfs.append(AdaBoostClassifier())
clfs.append(BaggingClassifier())
clfs.append(RandomForestClassifier())
clfs.append(GradientBoostingClassifier())
clfs.append(SVC())

for classifier in clfs:
    pipeline.set_params(clf = classifier)
    scores = cross_validate(pipeline, X_train, y_train)
    print(str(classifier))
    for key, values in scores.items():
            print(key,' mean ', values.mean())
```

```python
from sklearn.model_selection import GridSearchCV
pipeline.set_params(clf= SVC())
pipeline.steps
```

```python
cv_grid = GridSearchCV(SVC(), param_grid = {
    'kernel' : ['linear', 'rbf'],
    'C' : [0.1,1,10,100],
    'gamma' : [1,0.1,0.01,0.001]})

cv_grid.fit(X_train, y_train)
```

```python
cv_grid.best_params_
```

```python
cv_grid.best_estimator_
```

```python
from sklearn.metrics import accuracy_score
y_predict = cv_grid.predict(X_test)
accuracy = accuracy_score(y_test,y_predict)
print('Accuracy %.2f%%' % (accuracy*100))
```
